```python
import pandas as pd
```

```python
main_df = pd.read_csv('../csv/strongbox_top_100k_militaryfirstresponder_brcc.csv')
check_df1 = pd.read_csv('../csv/strongbox_top_150k_donorpropensity.csv')
check_df2 = pd.read_csv('../csv/strongbox_top_75k_donorpropensity_ca.csv')
```

```python
main_df = main_df[~main_df.individual_id.isin(check_df1.individual_id)]
```

```python
main_df = main_df[~main_df.individual_id.isin(check_df2.individual_id)]
```

```python
main_df.drop(columns=['rn'], inplace=True)
```

```python
main_df.to_csv('../csv/strongbox_top_100k_militaryfirstresponder_brcc.csv', index=False)
```
